```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [18.50, 13.50]
plt.rcParams["figure.autolayout"] = True
x = np.linspace(-10, 10, 50)
y = np.sin(x)
plt.axhline(y=0, c="green", linestyle="dashdot", label="y=0")
plt.plot(x, y, c="red", lw=2, linestyle="dashdot", label="y=sin(x)")
plt.grid(False)
plt.axis('off')
plt.legend()
plt.show()
```
